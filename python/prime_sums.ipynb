```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
#   ,    -  . 

import matplotlib as mpl
default_dpi = mpl.rcParamsDefault['figure.dpi']
mpl.rcParams['figure.dpi'] = default_dpi*1.5

with open('/Users/noob/primes_to10^6') as f:
    primes = f.readlines()
primes = [int(p) for p in primes if p]

sums = []
for i in primes:
    
    summ = sum([int(s) for s in str(i)])
    #if all([summ%j for j in range(2,summ)]):
    sums.append(summ)
print(sorted(set([i for i in range(2,max(sums)+1)]) - set(sums)))
    
    
plt.scatter(*zip(*enumerate(sums)), c=sums, s=1)
        
```
