```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


import matplotlib as mpl
default_dpi = mpl.rcParamsDefault['figure.dpi']
mpl.rcParams['figure.dpi'] = default_dpi*1.5

with open('../primes_to_10^6') as f:
    primes = f.readlines()
primes = [int(p) for p in primes if p]

sums = []
for i in primes:
    summ = sum([int(s) for s in str(i)])
    sums.append(summ)

plt.scatter(*zip(*enumerate(sums)), s=0.01)    

# list the gaps
print(sorted(set([i for i in range(2,max(sums)+1)]) - set(sums)))
```
